#### Решите систему уравнений:
#### y = x**2 – 1
#### exp(x) + x∙(1 – y) = 1

```python
from scipy.optimize import fsolve

def f(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x * (1 - y) - 1)

x1, y1 = fsolve(f, (-2, 5))
x2, y2 = fsolve(f, (2, 5))
x3, y3 = fsolve(f, (4, 15))

print('Решения системы уравнений:')
print(f'Первое решение: {x1}, {y1}')
print(f'Второе решение: {x2}, {y2}')
print(f'Третье решение: {x3}, {y3}')
```

#### Численное решение системы "вручную"

```python
a = 5.
n = 1000
x = np.linspace(-a, a, n)
y1 = (x**2 - 1)
y2 = (np.exp(x) + x - 1)/x

solution_x = []
solution_y = []

for i in range(n-1):
    if (y1[i] <= y2[i] <= y1[i+1]) | (y1[i] >= y2[i] >= y1[i+1]):
        solution_x.append(x[i])
        solution_y.append(y2[i])
        
for i in range(len(solution_x)):
    plt.plot(solution_x[i], solution_y[i], 'o', color='red')
plt.plot(x, x**2 - 1, color='blue', linewidth=1.0)
plt.plot(x, (np.exp(x) + x - 1)/x, color='green', linewidth=1.0)
plt.grid(True)
plt.title('Красными точками отмечены решения системы уравнений', color='red');
```

#### Решите систему уравнений и неравенств:
#### y = x**2 – 1
#### exp(x) + x∙(1 – y) > 1

```python
a = 5.
n = 100
x = np.linspace(-a, a, n)
y1 = (x**2 - 1)
y2 = ((np.exp(x) + x - 1)/x)

solution_x = []
solution_y = []

for i in range(n-1):
    if y1[i] <= y2[i]:
        solution_x.append(x[i])
        solution_y.append(y1[i])
        
plt.plot(x, (np.exp(x) + x - 1)/x, color='yellow', linewidth=3.0)        
for i in range(len(solution_x)):
    plt.plot(solution_x[i], solution_y[i], 'o', color='red')
plt.plot(x, x**2 - 1, color='blue', linewidth=1.0)
plt.grid(True)
plt.title('Красными точками отмечены решения системы в рамках интервала', color='red');
```

```python
print(f'Решения системы в рамках интервала по оси 0X от {-a} до {a} с точностью {n} точек на указанный интервал:')
for i in range(len(solution_x)):
    print(f'Решение номер {i}: {solution_x[i]}, {solution_y[i]}')
```

##### Дополните код Монте-Карло последовательности независимых испытаний расчетом соответствующих вероятностей (через биномиальное распределение) и сравните результаты. Повторите расчеты, взяв другие значения n и k.

```python
import itertools
from math import factorial as fct

k, n = 0, 10
a = np.random.randint(0, 2, n)
b = np.random.randint(0, 2, n)
c = np.random.randint(0, 2, n)
d = np.random.randint(0, 2, n)
x = a + b + c + d
for i in range(0, n):
    if x[i] == 2:
        k += 1
print(f'Фактическое выпадение удачи 2 раза подряд в 4 сериях из {n} испытаний = {k/n}')

kk = 2
nn = 4
koeff = fct(nn)/(fct(kk)*fct(nn - kk))
print('Расчётное значение по формуле Бернулли:')
print(f'Вероятность = (количество сочетаний из {nn} по {kk}) * (1/2^{nn}) = {koeff} * {1/(2**nn)} = {koeff/(2**nn)}')
```

##### Дополните код расчетом коэффициента корреляции x и y по формуле

$$ R = \frac{\sum((x_i - x_m)*(y_i - y_m))} {\sqrt(\sum(x_i - x_m)^2*\sum(y_i - y_m)^2)} $$

```python
n = 100
r = 0.7
x = np.random.rand(n)
y = r * x + (1 - r) * np.random.rand(n)
plt.plot(x, y, 'o')
plt.grid(True)
plt.show()
c = np.corrcoef(x, y)
print(f'Корреляция x и y = {c[0,1]}')
```
